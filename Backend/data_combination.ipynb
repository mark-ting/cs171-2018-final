```python
# 2018
months = ['01', '02', '03', '04', '05', '06',
         '07', '08', '09']
df_list = []
for i,val in enumerate(months):
    if i<3:
        df = pd.read_csv('2018'+val+'_hubway_tripdata.csv')
    elif i<4:
        df = pd.read_csv('2018'+val+'-hubway-tripdata.csv')
    else:
        df = pd.read_csv('2018'+val+'-bluebikes-tripdata.csv')
    
    df_list.append(df)
    
trip_2018 = pd.concat(df_list)
trip_2018.head()
```

```python
trip_2018.to_csv('hubway_Trips_2018.csv')
```
